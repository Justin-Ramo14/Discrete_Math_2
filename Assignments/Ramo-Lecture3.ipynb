```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_nodes_from([1, 2, 3, 4, 5])

# Add edges
edgelist = [(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)]
G.add_edges_from(edgelist)

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, arrowsize=20)
plt.title("Directed Graph")
plt.show()
```

3. This is a Complete graph since every vertex is connected to every other vertex

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a bipartite graph
kabit_network = nx.Graph()

# Add nodes with bipartite attribute
kabit_network.add_nodes_from(['Justin', 'Yza', 'Jasmine'], bipartite=0)  # Artists
kabit_network.add_nodes_from(['Mack', 'Niko'], bipartite=1)  # Songs

# Add edges between artists and songs
kabit_network.add_edges_from([('Justin', 'Mack'), ('Justin', 'Niko'),
                              ('Yza', 'Mack'), ('Yza', 'Niko'),
                              ('Jasmine', 'Mack'), ('Jasmine', 'Niko')])

# Set the positions of the nodes
top_nodes = {n for n, d in kabit_network.nodes(data=True) if d['bipartite'] == 0}
pos = nx.bipartite_layout(kabit_network, top_nodes)

# Draw the bipartite graph
nx.draw(kabit_network, pos=pos, with_labels=True, node_color=['pink', 'magenta', 'red', 'deepskyblue', 'cyan'])
plt.title("Bipartite Graph")
plt.show()
```

3. This is a Bipartite graph since  the vertices are divided into two disjoint sets such that all edges connect a vertex in one set to a vertex in another set

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes
G.add_nodes_from([1, 2, 3, 4])

# Add edges
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Create a subgraph
nodes_of_interest = [1, 2, 3]  # Specify the nodes you want in the subgraph
subgraph = G.subgraph(nodes_of_interest)

# Plot the original graph
plt.figure(figsize=(8, 4))
plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
plt.title("Original")
# Plot the subgraph
plt.subplot(122)
nx.draw(subgraph, with_labels=True, font_weight='bold')
plt.title("Subgraph")
plt.show()
```

4. The subgraph is like a smaller piece of the original graph, containing only certain selected dots and the lines connecting them, but it still keeps everything connected within that smaller group.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes
G.add_nodes_from([1, 2, 3, 4])

# Add edges to form an Eulerian circuit
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Check if the graph has an Eulerian circuit
if nx.is_eulerian(G):
    euler_circuit = list(nx.eulerian_circuit(G))
    print("Eulerian Circuit:", euler_circuit)
    
    # Plot the graph with the Eulerian circuit
    plt.figure(figsize=(6, 6))
    pos = nx.circular_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=1000)
    nx.draw_networkx_edges(G, pos, edgelist=euler_circuit, edge_color='r', width=2)
    plt.title("Graph with Eulerian Circuit")
    plt.show()
else:
    print("The graph does not have an Eulerian circuit.")
```

5. This is an Euler's circuit since every vertex has an even degree also is is a closed walk that traverses every edge of the graph exactly once  starting and ending at the same node

   Example: If you start from 1->4->3->2->1 notice that we only visted every edge once and we ended with the same vertex

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add edges to the graph
edges = [(1, 2), (2, 3), (3, 4), (4, 1), (1, 5), (5, 6), (6, 2), (2, 7), (7, 3), (3, 8), (8, 4)]
G.add_edges_from(edges)
plt.title("Eulers Trail")
# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```

6. This is an Eulers Trail since every degree of a vertex is 2 and also we can visit every edge exactly once.

     Example: 8->3->7->2->6->5->1->4->3->2->1

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a Hamiltonian circuit
G = nx.Graph()
nodes = ['A', 'B', 'C', 'D', 'E', 'F']
G.add_nodes_from(nodes)
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'F'), ('F', 'A')]
G.add_edges_from(edges)

# Draw the graph
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw_networkx_nodes(G, pos, node_size=700)
nx.draw_networkx_edges(G, pos, edgelist=edges, width=6)
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')

plt.axis('off')
plt.show()
```
