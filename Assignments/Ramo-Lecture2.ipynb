##### Question 1: Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$

To find the expected value $E[Y]$, we use the definition of expected value, which is the sum of each possible value of the random variable multiplied by its probability.

For this experiment, $Y_i$ represents the number of trials where the outcome falls into cell $i$. So, we have:

$$E[Y_i] = \sum_{y=0}^{n} y \cdot P(y)$$

Given that the probability distribution follows a multinomial distribution, we can substitute $P(y)$ with the multinomial probability mass function:

$$E[Y_i] = \sum_{y=0}^{n} y \cdot \frac{n!}{y_1!y_2!...y_n!} p_1^{y_1} p_2^{y_2}...p_n^{y_n}$$

Now, $y$ represents the number of trials where the outcome falls into cell $i$, which means $y_i = y$ and $y_j = 0$ for $j \neq i$. So, the multinomial term simplifies to:

$$\frac{n!}{y!(n-y)!} p_1^{y_1} p_2^{y_2}...p_n^{y_n} = \frac{n!}{y!(n-y)!} p_i^{y} (1-p_i)^{n-y}$$

Now, let's evaluate the sum:

$$E[Y_i] = \sum_{y=0}^{n} y \cdot \frac{n!}{y!(n-y)!} p_i^{y} (1-p_i)^{n-y}$$

Expanding this sum gives:

$$E[Y_i] = \sum_{y=0}^{n} \frac{n!}{(y-1)!(n-y)!} p_i^{y} (1-p_i)^{n-y}$$

$$E[Y_i] = \sum_{y=1}^{n} \frac{n!}{(y-1)!(n-y)!} p_i^{y} (1-p_i)^{n-y}$$

Now, we can notice that the sum looks like the expansion of $(p_i + (1-p_i))^n$, which equals 1. So, the sum actually equals 1:

$$E[Y_i] = 1 \cdot 1 = 1$$

Finally, since each $Y_i$ is identical in distribution, we have:

$$E[Y] = E[Y_1 + Y_2 + ... + Y_k] = E[Y_1] + E[Y_2] + ... + E[Y_k] = k \cdot E[Y_i] = k$$

Given that $k = n$ (the number of trials), we have:

$$E[Y] = n \cdot p_i$$

##### Question 4: Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective.

Given that the lot contains 5% defective fuses, the probability of selecting a defective fuse in one trial is $ p = 0.05 $. Therefore, the probability of selecting a non-defective fuse is $ q = 1 - p = 0.95 $.

We'll use the binomial distribution formula:

$$P(Y = k) = \binom{n}{k} \times p^k \times q^{n-k}$$

Where:
- $ n = 5 $ (the number of trials, i.e., the number of fuses tested)
- $ k $ represents the number of defective fuses observed
- $ p = 0.05 $ (the probability of selecting a defective fuse)
- $ q = 1 - p = 0.95 $ (the probability of selecting a non-defective fuse)

We want to find the probability of observing at least one defective fuse, so we'll calculate the probability of observing 0 defective fuses and then subtract it from 1.

Let's calculate:

$ P(Y = 0) = \binom{5}{0} \times (0.05)^0 \times (0.95)^{5-0} $

$ P(Y = 0) = \binom{5}{0} \times (0.05)^0 \times (0.95)^5 $

$ P(Y = 0) = 1 \times 1 \times (0.95)^5 $

$ P(Y = 0) \approx 1 \times 1 \times 0.7737809375 $

$ P(Y = 0) \approx 0.7737809375 $

Now, we'll find the probability of observing at least one defective fuse:

$ P(\text{At least one defective}) = 1 - P(Y = 0)$

$P(\text{At least one defective}) = 1 - 0.7737809375 $

##### Question 1: Find a general formula for the expected value of this distribution.
    To find the expected value (mean) of the probability distribution $ p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $, we can use the formula for the expected value of a discrete random variable:

$ E(Y) = \sum_{y=0}^{\infty} y \cdot p(y) $

Substituting the given probability distribution into this formula:

$ E(Y) = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} $

$ E(Y) = e^{-\lambda} \sum_{y=0}^{\infty} \frac{\lambda^y}{(y-1)!} $

$ E(Y) = e^{-\lambda} \left( \lambda \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!} \right) $

$ E(Y) = e^{-\lambda} \left( \lambda \sum_{y=0}^{\infty} \frac{\lambda^y}{y!} \right) $

$ E(Y) = e^{-\lambda} \left( \lambda e^{\lambda} \right) $

$ E(Y) = \lambda$

So, the general formula for the expected value of this distribution is $ \lambda $.
